# ETL
Se cargaran los datasets y posteriormente se realizaran sus respectivas modificaciones

```python
# Librerias a usar
import pandas as pd
import numpy as np
import matplotlib
```

```python
# Importacion de los datasets
df_Hechos = pd.read_csv('Data\Hechos_Homicidios.csv', sep= ",", )
df_Victimas = pd.read_csv("Data\Victimas_Homicidios.csv", sep= ",")
```

```python
# Vericacion de nulos por cada columna
df_Hechos.isnull().sum()
```

```python
# Eliminar filas con más de 3 valores nulos
df_Hechos_1 = df_Hechos[df_Hechos.isnull().sum(axis=1) <= 3]
```

```python
# Eliminacion de datos duplicados
df_Hechos_1 = df_Hechos_1.drop_duplicates()
```

Crearemos una nueva columna llamada "Momento del dia" la cual nos dira si el accidente ocurrio surante el dia o la noche 

```python
# Conversion del formato de la columna 'HH' a numerico 
df_Hechos_1['HH'] =  pd.to_numeric(df_Hechos_1['HH'], errors='coerce')

# Creamos la condicion para clasificar el dia o la noche 
condicion_dia = (df_Hechos_1['HH'] >= 6) & (df_Hechos_1['HH'] <= 18)

# Crear la nueva columna 'momento_del_dia' usando np.where
df_Hechos_1['Momento_del_dia'] = np.where(condicion_dia, 'Dia', 'Noche')

# Mostrar los resultados
df_Hechos_1
```

Se puede obrservar que hay un dato de la comuna '0' la cual no existe por que se arreglara buscando con su direccion o localizacion

```python
# Fila en donde el dato de la columa = 0
df_Hechos_1[df_Hechos_1['COMUNA'] == 0]
```

```python
# Fila con datos de la victima para el accidete '2016-0174'	
df_Victimas[df_Victimas['ID_hecho']=='2016-0174']
```

Al tener tantos datos nulos se toma la desicion de eliminar estas filas porque pueden tarernos problemas a la hora de hacer el analisis estadistico

```python
# Eliminacion de las filas 
df_Hechos_1.drop(df_Hechos_1[df_Hechos_1['COMUNA'] == 0].index, inplace=True)
df_Victimas.drop(df_Victimas[df_Victimas['ID_hecho']=='2016-0174'].index, inplace=True)
```

# Victimas

```python
# Vista previa del dataframe de Victimas
df_Victimas
```

```python
df_Victimas.info()
```

```python
df_Victimas.isnull().sum()
```

```python
# Eliminar de datos nulos 
df_Victimas_1 = df_Victimas[df_Victimas.isnull().sum(axis=1)<=1]
```

```python
# Eliminar datos duplicados
df_Victimas_1 = df_Victimas_1.drop_duplicates()
```

```python
# Transformar los datos de la columna 'EDAD' a numerico 
df_Victimas_1['EDAD'] =  pd.to_numeric(df_Victimas_1['EDAD'], errors='coerce')
```

```python
df_Victimas_1
```

```python
frecuencia = df_Victimas_1['AAAA'].value_counts().sort_index()
print(frecuencia)
```

```python
frecuencia = df_Victimas_1['MM'].value_counts().sort_index()
print(frecuencia)
```

```python
frecuencia = df_Victimas_1['SEXO'].value_counts().sort_index()
print(frecuencia)
```

```python
frecuencia = df_Victimas_1['EDAD'].value_counts().sort_index()
print(frecuencia)
```

```python
# Definir los rangos etarios
bins = [0, 18, 30, 50, float('inf')]
etiquetas = ['0-18', '19-30', '31-50', '51+']

# Crear una nueva columna 'RangoEtario' utilizando pd.cut
df_Victimas_1['RangoEtario'] = pd.cut(df_Victimas_1['EDAD'], bins=bins, labels=etiquetas, right=False)

# Mostrar los resultado
df_Victimas_1
```

```python
frecuencia = df_Victimas_1['RangoEtario'].value_counts().sort_index()
print(frecuencia)
```

Una vez hecho las respectivas modificaciones como fueron las eliminaciones de nulos, conversion de formatos y verifaciones de valores duplicados se procede a cargar los dataframes para realizar su analisis estadistico

```python
# Exportacion de los dataframe en formato parquet
df_Hechos_1.to_parquet('Data\Hechos_Homicidios_ETL.parquert', index = False)
df_Victimas_1.to_parquet('Data\Victimas_Homicidios_ETL.parquert', index = False)
```
